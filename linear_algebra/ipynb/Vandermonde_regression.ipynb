# Generate $x$ and $y$

```python
#vjgdruihdiruhgdiuhj
n=10
x = np.arange(1,n+1,1)
x = x/n  
y = np.random.randn(n)*10+30
```

```python
plt.plot(x,y)
```

# Vandermonde matrix

```python
A = np.vander(x,increasing=True)
```

```python
plt.imshow(A)
```

### Compute the inverse of $A$

```python
A_inv = np.linalg.inv(A)
```

```python
plt.imshow(A_inv )
```

```python
plt.imshow(A.dot(A_inv))
```

### Compute $c$

```python
c = A_inv.dot(y)
```

```python
plt.plot(c)
```

```python
plt.plot(A.dot(c))
```
